## Q1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
def HEADER_TAGS(url):
# By inspecting the page we find out that there are only 3 header tags.So  creating 3 empty list.
    h1=[]
    h2=[]
    h3=[]

    # Getting page permission
    page=requests.get(url)
    wikipedia_soup=BeautifulSoup(page.content,"html.parser")

    # First header
    header_h1=wikipedia_soup.find_all("h1")
    for i in header_h1:
        h1.append(i.string)

    # Second header
    header_h2=wikipedia_soup.find_all("h2")
    for i in header_h2:
        h2.append(i.string or i.text)

    # Third header    
    header_h3=wikipedia_soup.find_all("h3")
    for i in header_h3:
        h3.append(i.text.replace("\n",""))
        
    # Printing all the header tags
    print("First Header tag - ",h1)
    print("\n")
    print("Second Header tag - ",h2)
    print("\n")
    print("Third Header tag - ",h3)
    
```

```python
# calling function
tags=HEADER_TAGS("https://en.wikipedia.org/wiki/Main_Page")
tags
```

## Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
# Creating function
def imdb_top_movies(url):
    
    # Creating empty list
    Movie_Name=[]
    IMDb_ratings=[]
    Release_Year=[]
 
    # Getting webpage permission
    imdb=requests.get(url)
    imdb_soup=BeautifulSoup(imdb.content,'html.parser')
    # For the name of movie
    name=imdb_soup.find_all("td",class_="titleColumn")
    for i in name:
        Movie_Name.append(i.find('a').get_text())
    # For the ratings of movie
    rating=imdb_soup.find_all("td",class_="ratingColumn imdbRating")
    for i in rating:
        IMDb_ratings.append(i.find("strong").get_text())
    # for the year of movie release
    year=imdb_soup.find_all("span",class_="secondaryInfo")
    for i in year:
        Release_Year.append(i.text.replace("(","").replace(")",""))
    
    # Creating dataframe
    movies=pd.DataFrame({})
    movies['Name']=Movie_Name
    movies['IMDB_Ratings']=IMDb_ratings
    movies['Year_of_Release']=Release_Year
    
    # Selecting top 100 movies
    movies=movies[:100]
    
    # Saving into csv file
    movies.to_csv("IMDb Top 100 rated movies.csv")
    
    return movies
```

```python
Movies=imdb_top_movies("https://www.imdb.com/chart/top/")
Movies
```

## Q4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
# Creating a function to find the book name,author , genere, and book review

def book(url):
    # Creating empty lists
    book_name = []
    author_name = []
    genre = []
    book_review = []
    
    # Getting webpage permission
    page = requests.get(url)
    book_soup=BeautifulSoup(page.content,"html.parser")
    # for the name of book
    name=book_soup.find_all("h4",class_="italic")
    for i in name:
        book_name.append(i.find("a").text.replace("\n",""))
    # for the authors of book
    author=book_soup.find_all("p",class_="sans bold")
    for i in author:
        author_name.append(i.text.replace("\n",""))
    # for book genre
    Genre=book_soup.find_all("p",class_="genre-links hidden-phone")
    for i in Genre:
        genre.append(i.text.replace("\n",""))
    # for book reviews
    reviews=book_soup.find_all("p",class_="excerpt")
    for i in reviews:
        book_review.append(i.text.replace("\n",""))
    
    # Creating a dataframe
    import pandas as pd
    books=pd.DataFrame({})
    books['Book_name']=book_name
    books['Author']=author_name
    books['Genre']=genre
    books['Reviews']=book_review
    books=books.head()
    return(books)
```

# Q5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
### i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def odi_top10(url):
    name=[]
    matches=[]
    points=[]
    ranking=[]
    
    # Getting the webpage request
    page=requests.get(url)
    odi_soup=BeautifulSoup(page.content,"html.parser")
    
    # Team name
    team=odi_soup.find_all("span",class_="u-hide-phablet")
    for i in team:
        name.append(i.text)
    
   
    # Banner contains differ class attributes than regular rankings. So taking them separately
    # Banner team matches
    matches.append(odi_soup.find("td",class_="rankings-block__banner--matches").text)
    
    # Other team matches
    match=odi_soup.find_all("td",class_="table-body__cell u-center-text")
    for i in range(0,len(match),2):
        matches.append(match[i].text)
            
    
    # Banner contains different class attribute in points also.
    # Banner team points
    point_ban=odi_soup.find("td",class_="rankings-block__banner--points")
    points.append(point_ban.text)
    
    # Other team points
    point=odi_soup.find_all("td",class_="table-body__cell u-center-text")
    for i in range(1,len(point),2):
        points.append(point[i].text)
        
    # Banner team ratings
    ranking.append(odi_soup.find("td",class_="rankings-block__banner--rating u-text-right").text.replace("\n","").replace(' ',''))
    
    # Other team ratings
    RATING=odi_soup.find_all("td",class_="table-body__cell u-text-right rating")
    for i in RATING:
        ranking.append(i.text)
        
    # Creating dataframe
    Rankings=pd.DataFrame({})
    Rankings['Team']=name[:10]
    Rankings['Matches']=matches[:10]
    Rankings['Runs_scored']=points[:10]
    Rankings['Ranking']=ranking[:10]
    
    return Rankings
```

```python
Men_top10=odi_top10("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
Men_top10
```

### Q5  ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
def odi_10_batsman(url):
    players=[]
    teams=[]
    ratings=[]
    
    # Getting page request
    page=requests.get(url)
    batsman_soup=BeautifulSoup(page.content,"html.parser")
    
    # Cricketer name
    # As there is banner present which has differnt class and attributes. We will treat them separately
    # Banner cricketer name
    BATSMAN=batsman_soup.find("div",class_="rankings-block__banner--name-large")
    players.append(BATSMAN.text.replace("\n",""))
    
    # Other cricketer's name
    batsman=batsman_soup.find_all("td",class_="table-body__cell rankings-table__name name")
    for i in batsman:
        players.append(i.text.replace("\n",""))
    players=players[:10]
        
        
    # Team
    # Banner team
    TEAM=batsman_soup.find("div",class_="rankings-block__banner--nationality")
    teams.append(TEAM.text.replace("\n","")) 
    
    # Other teams
    team=batsman_soup.find_all("span",class_="table-body__logo-text")
    for i in team:
        teams.append(i.text.replace("\n",""))
    teams=teams[:10]
    
    
    # Rating
    # Banner rating
    RATING=batsman_soup.find("div",class_="rankings-block__banner--rating")
    ratings.append(RATING.text)
    
    # Other cricketer ratings
    rating=batsman_soup.find_all("td",class_="table-body__cell rating")
    for i in rating:
        ratings.append(i.text)
    ratings=ratings[:10]
    
    
    # Creating dataframe
    Batsman_ranking=pd.DataFrame()
    Batsman_ranking['Player']=players
    Batsman_ranking['Team']=teams
    Batsman_ranking['Rating']=ratings
        
    return Batsman_ranking
```

## Q7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def amazon_mobiles(url):
    # Creating emopty list
    product_name = []
    price = []
    avg_ratings=[]
    image_url=[]
    
    # Getting page request
    page=requests.get(url)
    amazon_soup=BeautifulSoup(page.content)

    product=amazon_soup.find_all("span",class_="a-size-medium a-color-base a-text-normal")
    for i in product:
        product_name.append(i.get_text())
    product_price=amazon_soup.find_all("span",class_="a-price-whole")
    for i in product_price:
        price.append(i.text)
    ratings=amazon_soup.find_all("span",class_="a-icon-alt")
    for i in ratings:
        avg_ratings.append(i.text.replace("[","").replace("]",""))
    image=amazon_soup.find_all("img",class_="s-image")
    for i in image:
        image_url.append(i.get("src"))
        
    # Creating a dataframe
    mobiles=pd.DataFrame()
    mobiles['Product']=product_name[:15]
    mobiles['Price']=price[:15]
    mobiles['Ratings']=ratings[:15]
    mobiles['Image_url']=image_url[:15]
    
    return mobiles
```

## Q8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def weather_sanfranscisco(url):
    period=[]
    short_description=[]
    temperature=[]
    description=[]
    
    # Get page request
    page=requests.get(url)
    weather_soup=BeautifulSoup(page.content,"html.parser")
    
    weather_period=weather_soup.find_all("p",class_="period-name")
    # For weather period
    for i in weather_period:
        if i.text==None:
            period.append(None)
        else:
            period.append(i.text.replace("\n",""))
    # Short description of weather
    descr=weather_soup.find_all("p",class_="short-desc")
    for i in descr:
        if i.text==None:
            short_description.append(None)
        else:
            short_description.append(i.text.replace("\n",""))
    # Temperature   
    temp=weather_soup.find_all("p",class_=["temp temp-high","temp temp-low"])
    for i in temp:
        if i.text==None:
            temperature.append(None)
        else:
            temperature.append(i.text.replace("\n","").split(":")[1])
    # Weather details description    
    descriptions=weather_soup.find_all("div",class_="col-sm-10 forecast-text")
    for i in descriptions:
        if i.text==None:
            description.append(None)
        else:
            description.append(i.text.replace("\n","").split("with a")[1])
        
    # Creating a dataframe
    weather=pd.DataFrame()
    weather['Period']=period[:9]
    weather['Short_Description']=short_description[:9]
    weather['Temperature']=temperature[:9]
    weather['Description']=description[:9]
    
    return weather
```
